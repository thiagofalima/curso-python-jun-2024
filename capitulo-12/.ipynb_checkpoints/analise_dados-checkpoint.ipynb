```python
import pandas as pd
```

```python
df = pd.read_csv('vgsales.csv')
```

```python
display(df)
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df['Year'] = [int(ano) for ano in df['Year']]
```

```python
display(df)
```

```python
df = df.drop(['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'], axis=1)
```

```python
display(df)
```

```python
jogos_2000 = df.loc[lambda tabela: tabela['Year'] == 2000]
```

```python
display(jogos_2000)
```

```python
jogos_2000 = jogos_2000.sort_values(by=['Global_Sales'], ascending=False)
```

```python
display(jogos_2000)
```

```python
top_5_jogos_2000 = jogos_2000.head()
```

```python
display(top_5_jogos_2000)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
```

```python
grafico = ax.bar(top_5_jogos_2000['Name'], top_5_jogos_2000['Global_Sales'], width=0.5)
ax.set_title('Top 5 Jogos mais vendidos de 2000')
ax.set_xlabel('Jogos')
ax.set_ylabel('Vendas Internacionais')
ax.bar_label(grafico, padding=0.3)
plt.show()
```
